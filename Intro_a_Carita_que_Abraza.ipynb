```python
# Use a pipeline as a high-level helper
from transformers import pipeline

traductor = pipeline("translation", model="Helsinki-NLP/opus-mt-es-en")
```

```python
traductor('Bienvenidos a la clase de hoy')[0]['translation_text']
```

```python
def traducir(oracion):
  return traductor(oracion)[0]['translation_text']
```

```python
traducir('Hola a todos!')
```

```python
traducir('Bonjour a tous~')
```

```python
traducir('Hallo Leute')
```
